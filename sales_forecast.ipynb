```python
import numpy as np
```

```python
# Product categories
products = np.array(["Foundations", "Lipsticks", "Face Creams", "Perfumes", "Body Lotions"])

# Monthly sales data (units sold or revenue in thousands)
sales = np.array([420, 650, 300, 500, 250])

print("Products:", products)
print("Original Sales:", sales)
```

```python
np.random.seed(42)  # for reproducibility

# Random variation between -50 and +60
variation = np.random.randint(-50, 60, size=5)

adjusted_sales = sales + variation

print("Market Variation:", variation)
print("Adjusted Sales:", adjusted_sales)
```

```python
growth_rate = 0.10

forecast_sales = adjusted_sales * (1 + growth_rate)

print("Forecast Sales:", forecast_sales)
```

```python
total_sales = adjusted_sales.sum()
average_sales = adjusted_sales.mean()
std_dev = adjusted_sales.std()

best_product = products[np.argmax(adjusted_sales)]
worst_product = products[np.argmin(adjusted_sales)]

print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
print("Standard Deviation:", std_dev)
print("Best Performer:", best_product)
print("Lowest Performer:", worst_product)
```

```python
print("=== Monthly Sales Summary ===\n")

for product, current, forecast in zip(products, adjusted_sales, forecast_sales):
    print(f"{product}: Current Sales = {current}, Forecast = {forecast:.2f}")

print("\n--- Statistical Overview ---")
print(f"Total Sales: {total_sales}")
print(f"Average Sales: {average_sales:.2f}")
print(f"Sales Standard Deviation: {std_dev:.2f}")
print(f"Best Performer: {best_product}")
print(f"Lowest Performer: {worst_product}")
```

```python
import matplotlib.pyplot as plt
```

```python
# Set bar positions
x = np.arange(len(products))  # number of products
width = 0.35  # width of bars

# Create figure
plt.figure()

# Plot current sales
plt.bar(x - width/2, adjusted_sales, width, label="Current Sales")

# Plot forecasted sales
plt.bar(x + width/2, forecast_sales, width, label="Forecast Sales")

# Labels and formatting
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Current vs Forecasted Sales - Beauty Brand")
plt.xticks(x, products)
plt.legend()

# Show plot
plt.show()
```
